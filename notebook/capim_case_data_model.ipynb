```python
# check if is not a list
non_lists = df_clinics[~df_clinics['INTEREST_REASONS'].apply(lambda x: isinstance(x, list))]
non_lists.columns
```

```python
def convert_to_list(x):
    try:
        return ast.literal_eval(x) if isinstance(x, str) else x
    except:
        return []

df_clinics_ = df_clinics.copy()
df_clinics_['INTEREST_REASONS'] = df_clinics['INTEREST_REASONS'].apply(convert_to_list)
```

```python
# replace NaN to empty list
df_clinics_['INTEREST_REASONS'] = df_clinics_['INTEREST_REASONS'].apply(lambda x: x if isinstance(x, list) else [])

# check NaN
df_clinics_['INTEREST_REASONS'].isna().sum()
```

```python
# check maximum list size
max_length = df_clinics_['INTEREST_REASONS'].apply(len).max()

# expand list
df_expanded = pd.DataFrame(df_clinics_['INTEREST_REASONS'].tolist(),
                           columns=[f'REASON_{i+1}' for i in range(max_length)])

df_clinics_ = pd.concat([df_clinics_, df_expanded], axis=1)

df_clinics_.head()
```

## Join tables
